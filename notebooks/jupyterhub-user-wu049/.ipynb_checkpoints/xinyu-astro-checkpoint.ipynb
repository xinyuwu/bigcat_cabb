```python
import matplotlib.pyplot as plt
import numpy as np

from astropy.visualization import astropy_mpl_style, quantity_support


import astropy.units as u
from astropy.coordinates import AltAz, EarthLocation, SkyCoord
from astropy.time import Time


plt.style.use(astropy_mpl_style)
quantity_support()


m33 = SkyCoord.from_name('M33')

time = Time('2023-03-09 01:00:00', scale='utc')
bear_mountain = EarthLocation(lat=41.3*u.deg, lon=-74*u.deg, height=390*u.m)
```

```python
time_interval = np.linspace(10, 28, 100)*u.hour

time_frame = time + time_interval
data_frame = AltAz(obstime=time_frame, location=bear_mountain)
m33altazs = m33.transform_to(data_frame)
plt.ylim(0*u.deg, 90*u.deg)
plt.plot(time_interval, m33altazs.alt, color='r', label='m33')
```
